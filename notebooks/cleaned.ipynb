```python
import pandas as pd
import os

# 1. Automatic path positioning (Ensures reliability across different environments)
# Get the absolute path of the current notebook
current_path = os.getcwd() 
# Get the project root directory (one level up from 'notebooks')
project_root = os.path.dirname(current_path) 

# Define absolute paths for raw data and processed folder
raw_data_path = os.path.join(project_root, 'data', 'raw', 'YRBS_2007.csv')
processed_folder = os.path.join(project_root, 'data', 'processed')

# 2. Create directory if it doesn't exist
os.makedirs(processed_folder, exist_ok=True)

# 3. Data Loading and Cleaning
df = pd.read_csv(raw_data_path)

# Drop missing values for height analysis
df_clean = df.dropna(subset=['HowTallAreYouWithoutShoesInMeters']).copy()

# Filter for valid smoking responses (1=Yes, 2=No)
df_clean = df_clean[df_clean['EverCigaretteUse'].isin([1, 2])].copy()

# Recode smoking variable: Replace 2 (No) with 0 for binary analysis (0=No, 1=Yes)
df_clean['EverCigaretteUse_Recoded'] = df_clean['EverCigaretteUse'].replace({2: 0})

# 4. Save processed data to data/processed
output_path = os.path.join(processed_folder, 'yrbs_cleaned.csv')
df_clean.to_csv(output_path, index=False)

print(f"--- Execution Summary ---")
print(f"✅ Data successfully loaded from: {raw_data_path}")
print(f"✅ Cleaned data saved to: {output_path}")
print(f"Final valid sample size (n): {len(df_clean)}")
```
