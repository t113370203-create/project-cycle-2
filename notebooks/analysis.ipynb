```python
import numpy as np
from scipy import stats 
from statsmodels.stats.proportion import proportion_confint, proportions_ztest
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
cleaned_data_path = '../data/processed/yrbs_cleaned.csv'

if os.path.exists(cleaned_data_path):
    df = pd.read_csv(cleaned_data_path)
    print(f"✅ Cleaned data loaded successfully!")
    print(f"Total Sample Size: {len(df)}")
    
    # --- TASK 1: Point Estimates for Proportion (Smoking) --- [cite: 98, 99]
    n_prop = len(df)
    x_prop = int(df['EverCigaretteUse_Recoded'].sum())
    p_hat = x_prop / n_prop

    # --- TASK 2: Point Estimates for Mean (Height) --- [cite: 104, 105]
    heights = df['HowTallAreYouWithoutShoesInMeters']
    n_mean = len(heights)
    x_bar = heights.mean()
    s_std = heights.std()

    print("\n=== Task 1 & 2: Point Estimates ===")
    print(f"[Smoking Proportion] n={n_prop}, Successes x={x_prop}, Sample Proportion (p-hat)={p_hat:.4f}")
    print(f"[Average Height] n={n_mean}, Mean (x-bar)={x_bar:.4f}, Std Deviation (s)={s_std:.4f}")
    
else:
    print(f"❌ File not found. Please check the path: {os.path.abspath(cleaned_data_path)}")
```

```python
# --- TASK 3: 95% Confidence Intervals (CI) ---

# 1. Compute Confidence Interval for Proportion (Smoking) [cite: 18, 101]
lower_p, upper_p = proportion_confint(count=x_prop, nobs=n_prop, alpha=0.05, method='normal')

# 2. Compute Confidence Interval for Mean (Height) [cite: 18, 107]
lower_m, upper_m = stats.t.interval(confidence=0.95, df=n_mean-1, loc=x_bar, scale=s_std/np.sqrt(n_mean))

print("=== Task 3: 95% Confidence Intervals ===")
print(f"Smoking Proportion 95% CI: ({lower_p:.4f}, {upper_p:.4f})")
print(f"Average Height 95% CI: ({lower_m:.4f}, {upper_m:.4f})")
```

```python
# --- TASK 4: Hypothesis Testing (p-values) ---

# 1. Proportion Test (Benchmark p0 = 0.50)
z_stat, p_val_p = proportions_ztest(count=x_prop, nobs=n_prop, value=0.50)

# 2. Mean Test (Benchmark mu0 = 1.70)
t_stat, p_val_m = stats.ttest_1samp(heights, popmean=1.70)

print("=== Task 4: One-Sample Test Results ===")
print(f"Smoking Proportion Test (vs 0.50) p-value: {p_val_p:.4e}")
print(f"Mean Height Test (vs 1.70) p-value: {p_val_m:.4e}")

# Statistical Decision Logic
def check_sig(p):
    return "Significant Difference (Reject H0)" if p < 0.05 else "No Significant Difference (Fail to Reject H0)"

print(f"\nStatistical Decisions:")
print(f"Smoking Proportion: {check_sig(p_val_p)}")
print(f"Student Height: {check_sig(p_val_m)}")
```
