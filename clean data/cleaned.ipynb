```python
import pandas as pd
import os

# 1. 讀取原始資料
# 根據你的路徑，如果檔案在 stat cycle 2 資料夾下
raw_data = 'YRBS_2007.csv'
df = pd.read_csv(raw_data)

# 2. 定義研究問題需要的變數 [cite: 28, 45]
# 比例分析: EverCigaretteUse (曾吸過菸)
# 平均數分析: HowTallAreYouWithoutShoesInMeters (身高)

# 3. 資料清洗：排除缺失值與無效代碼 [cite: 68]
# 只保留有身高數據的人
df_clean = df.dropna(subset=['HowTallAreYouWithoutShoesInMeters']).copy()

# 只保留吸菸代碼為 1 (Yes) 或 2 (No) 的人 
df_clean = df_clean[df_clean['EverCigaretteUse'].isin([1, 2])].copy()

# 4. 重編碼 (Recoding Rules) [cite: 34, 69]
# 關鍵：將 Code 2 (Failure) 轉為 0，這樣 sum() 才會是成功人數
df_clean['EverCigaretteUse_Recoded'] = df_clean['EverCigaretteUse'].replace({2: 0})

# 5. 檢查資料狀況 [cite: 70, 78]
n_final = len(df_clean)
x_success = df_clean['EverCigaretteUse_Recoded'].sum()
p_hat = x_success / n_final

print(f"--- 資料檢查報告 ---")
print(f"原始資料筆數: {len(df)}")
print(f"最終有效樣本數 (n): {n_final}") 
print(f"曾吸菸的成功人數 (x): {x_success}") 
print(f"樣本比例 (p-hat): {p_hat:.4f}") # 這次一定會介於 0~1 之間 [cite: 99]

# 6. 儲存處理後的資料 [cite: 144, 167]
if not os.path.exists('processed'):
    os.makedirs('processed')
df_clean.to_csv('processed/yrbs_cleaned.csv', index=False)
print("\n✅ 清洗完成，檔案已儲存至 processed/yrbs_cleaned.csv")
```
