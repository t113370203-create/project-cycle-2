```python
# 讀取清洗後的資料
df = pd.read_csv('processed/yrbs_cleaned.csv')

# --- TASK 1: 比例點估計 (吸菸) ---
n_prop = len(df)
x_prop = int(df['EverCigaretteUse_Recoded'].sum())
p_hat = x_prop / n_prop

# --- TASK 2: 平均數點估計 (身高) ---
heights = df['HowTallAreYouWithoutShoesInMeters']
n_mean = len(heights)
x_bar = heights.mean()
s_std = heights.std()

print("=== Task 1 & 2: Point Estimates ===")
print(f"【吸菸比例】樣本數 n={n_prop}, 成功次數 x={x_prop}, 樣本比例 p-hat={p_hat:.4f}")
print(f"【平均身高】樣本數 n={n_mean}, 平均數 x-bar={x_bar:.4f}, 標準差 s={s_std:.4f}")
```

```python
# --- TASK 3: Confidence Intervals (95%) ---

# 比例的信賴區間
ci_p = proportion_confint(count=x_prop, nobs=n_prop, alpha=0.05, method='normal')

# 平均數的信賴區間 (使用 t 分佈)
ci_m = stats.t.interval(0.95, df=n_mean-1, loc=x_bar, scale=s_std/(n_mean**0.5))

print("=== Task 3: 95% Confidence Intervals ===")
print(f"曾吸菸比例的 CI: [{ci_p[0]:.4f}, {ci_p[1]:.4f}]")
print(f"平均身高的 CI: [{ci_m[0]:.4f}, {ci_m[1]:.4f}]")
```

```python
# --- TASK 4: Hypothesis Testing (p-values) ---

# 1. 比例檢定 (基準值 p0 = 0.50)
z_stat, p_val_p = proportions_ztest(count=x_prop, nobs=n_prop, value=0.50)

# 2. 平均數檢定 (基準值 mu0 = 1.70)
t_stat, p_val_m = stats.ttest_1samp(heights, popmean=1.70)

print("=== Task 4: One-Sample Test Results ===")
print(f"吸菸比例檢定 (vs 0.50) p-value: {p_val_p:.4e}")
print(f"身高平均檢定 (vs 1.70) p-value: {p_val_m:.4e}")

# 簡單判定邏輯
def check_sig(p):
    return "顯著差異 (拒絕 H0)" if p < 0.05 else "無顯著差異 (不拒絕 H0)"

print(f"\n結論判定：")
print(f"吸菸比例：{check_sig(p_val_p)}")
print(f"學生身高：{check_sig(p_val_m)}")
```
